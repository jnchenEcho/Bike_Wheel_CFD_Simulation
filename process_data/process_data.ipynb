```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')  # Use default style for better visibility
```

```python
# 创建三组数据
data = {
    "40mm": np.column_stack((
        np.linspace(30, 60, 7),    # speed
        np.array([0.014395908,0.019338864,0.024992452,0.031376693,0.038477967,0.046298717,0.054836598])  # Cd
    )),
    "50mm": np.column_stack((
        np.linspace(30, 60, 7),
        np.array([0.011658389,0.015655317,0.02022828,0.025391064,0.031134827,0.037457324,0.044355516])
    )),
    "60mm": np.column_stack((
        np.linspace(30, 60, 7),
        np.array([0.009956988,0.013368116,0.017267598,0.021668232,0.026567205,0.031955448,0.037842912])
    ))
}
```

```python
# 绘图

plt.figure(figsize=(8,5))

styles = {
    "40mm": {"color": "tab:blue",  "marker": "o", "linewidth": 2.0, "markersize": 6},
    "50mm": {"color": "tab:orange","marker": "s", "linewidth": 2.5, "markersize": 6},
    "60mm": {"color": "tab:green", "marker": "^", "linewidth": 3.0, "markersize": 6},
}

for label, arr in data.items():
    style = styles[label]
    speed = arr[:,0]
    cd = arr[:,1]
    plt.plot(speed, cd, linestyle='-', label=label, **style)

plt.xlabel("Speed (m/s)", fontsize=12)
plt.ylabel("Cd", fontsize=12)
plt.title("Speed vs Cd for Different Rim Depths", fontsize=14)
plt.legend(title="Rim Depth")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
